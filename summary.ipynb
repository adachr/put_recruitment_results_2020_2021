```python
file_path = os.path.join(os.getcwd(), "main.csv")
df = pd.read_csv(file_path)
df
```

#### Rozkład płci ogółem

```python
print(df['Płeć'].value_counts().to_string())

df['Płeć'].value_counts().plot.pie(autopct='%1.1f%%', colors=['C0', 'C1'])
plt.show()
```

```python
# print(df.groupby('Wydział')['Płeć'].value_counts().to_string())
# df.groupby('Wydział')['Płeć'].value_counts().aggregate(np.median).reset_index().sort_values('Płeć').unstack().plot.barh(color=['C1', 'C0'])
df.groupby('Wydział')['Płeć'].value_counts(normalize=True).unstack().plot.barh(color=['C1', 'C0'])

# plt.show()
```

```python
df.groupby('Wydział')['Płeć'].value_counts().sort_index()
```

```python
df['Liczba punktów'].value_counts(bins=10, ascending=True)
```

#### Ogólny rozkład puntków dla studentów 1 stopnia

```python
bins = pd.cut(df.loc[df['Stopień'] == 1, 'Liczba punktów'], bins=np.linspace(0, 1000, 41))
tb = df.loc[df['Stopień'] == 1].groupby(bins)['Liczba punktów'].agg(['count'])
tb[tb['count'] !=0].plot.bar(figsize=(10,5))
```
